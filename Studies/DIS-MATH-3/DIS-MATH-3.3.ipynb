**Задача 13.**  
Сколько всего неизоморфных подграфов, порождённых четырьмя вершинами, у приведённого ниже графа?



Немного не понятно было сначала, что конкретно спрашивается - количество возможных сочетаний или видов (судя по ответу - именно видов). Сочетаний из 4-х вершин в данном графе, если вспомнить комбинаторику, в данном случае используя формулу неупорядоченных сочетаний без повторений - 7!/4!(7-4)! = 210/6 = 35. Но так как в задаче имеется в виду количество типов графов, и вычленять все графы не нужно, отметим те, которые присутствуют из возможных видов графа. Возможные виды - это графы с 4 ребрами (каждая вершина в степени 2, представляет собой квадрат), 5 ребрами (2 вершины 2 степени и 2 - 3 степени), и полные графы с 6 ребрами (каждая вершина имеет 3 степень). Видим, что граф с 4 ребрами - 1-2-4-7; с 5 ребрами - 1-2-3-4; полный граф - 1-2-3-5. То есть, в данном графе образуются все 3 типа порождённых подграфов с 4 вершинами. **Итоговый ответ: 3.**

**Задача 14.**  
Предложенная разработчиками рекомендательная система некоторой социальной сети довольно агрессивна и предлагает пользователей друг другу, как потенциальных друзей, если у них есть хотя бы один общий друг. Её решили опробовать на небольшой выборке. Рекомендации были разосланы всем одновременно и лишь раз. Известно, что 10 предложений не были одобрены. Сколько образовалось новых пар-друзей?



Тут просто нужно соединить ребрами те пары, которые до этого не были друзьями, но у которых есть общие друзья, хотя бы один. Из этого можно составить табличку с новыми парами, но из-за громозкости решения я просто напишу сводку, то есть сколько пар (новых!) образовалось для конкретной вершины: 1-3; 2-5; 3-2; 4-4; 5-3; 6-3; 7-4; 8-5; 9-2; 10-3. Для подсчёта количества потенциальных пар, нужно сложить все полусумму (так как убираем повторы, например, рекомендацию типа 1-7 и 7-1 считаем как одну) всех новых пар. Итого у нас получится (3+5+2+4+3+3+4+5+2+3)/2=34/2=17 предложений. 10 предложений были отклонены, итого осталось: 17-10=7 предложений были приняты, что и является ответом.

Задача похожа на предыдущую, за исключением того, что общих друзей должно быть 2 минимум, также сказано, что запрос не один, а раз в день на протяжении недели (или "до талого", пока не останется возможных вариантов). **Задача 23.**  
Вы получили заказ от фирмы-туроператора: от вас требуется составить туристический маршрут, проходящий по указанным объектам и улицам. Вам предъявлены следующие требования относительно улиц: улицы ведущие к дворцу
(A), как и улицы ведущие к главной площади (B) объявлены ключевыми и обязаны входить в маршрут; также никакая улица не может дважды входить в маршрут. При этом желательно охватить как можно больше улиц, а именно — они предлагают оплату по следующей схеме: будет выплачено S/(k+1) р., где S=100000 р., а k — количество непосещённых улиц в предложенном маршруте.



Какова максимальная сумма, которую можно заработать по этому контракту? Предложите маршрут, при котором она достигается.

Для того, чтобы составить требуемый маршрут, нужно найти эйлеров путь. Для этого должно выполниться условие: развилок с нечетным количеством путей (рёбер) должно быть 2 или 0. У нас их 4. Можно удалять развилки, но так, чтобы прибыль была максимальной, за каждую удалённую развилку, судя по формуле, она уменьшается. После внимательного осмотра удаляем путь 2-4. Итого нечетных развилок остаётся 2, это неприкасаемые А и В. Далее просто построим эйлеров путь, выполняя условия задачи. У меня получился путь A -> 1 -> B -> 7 -> 4 -> 1 -> 3 -> 5 -> 6 -> 2 -> A -> 3 -> 6 -> 7 -> 5 -> B. Удалили 1 ребро, и сумма выплаты за составление маршрута составит 100000/(1+1) = 50000р.
