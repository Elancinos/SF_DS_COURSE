### Юнит 3. Перестановки

Перестановки без повторений:

```python
# Количество букв 4 
n = 4

# перестановки, так как важен порядок и буквы не возвращаются в выборку
permutations = factorial(n)
permutations
```

Перестановки с повторениями:

```python
word = "ИНТЕЛЛЕКТ"
n = len(word)

# Посчитаем, сколько раз повторяется каждая буква
letter_counts = Counter(word)

# Вычислим знаменатель — произведение факториалов количеств повторений
denominator = 1
for count in letter_counts.values():
    denominator *= factorial(count)

# Общее количество уникальных перестановок
total_permutations = factorial(n) // denominator

print(f"Количество уникальных перестановок: {total_permutations}")
```

### Юнит 4. Размещения

Размещения без повторений:

```python
n = 10  # общее количество аналитиков
k = 3   # количество выбираемых аналитиков

'''Используя формулу размещений и функцию math.factorial() рассчитаем количество возможных вариантов'''
arrangements = factorial(n) // factorial(n - k)

# Вывод результата
arrangements
```

Размещения с повторениями:

```python
n = 10  # количество различных цифр
k = 4   # длина PIN-кода

# Вычисляем размещение с повторениями
result = n ** k

# Вывод результата
result
```

```python
# Простая реализация обработки коллизий из принципа Дирихле

class HashTable:
    '''Конструктор класса, который
    принимает размер хеш-таблицы (size) и инициализирует таблицу (self.table) списком None указанного размера.'''
    def __init__(self, size):
        self.size = size
        self.table = [None] * size
        
    '''Функция, которая определяет метод, реализующий простую хеш-функцию для строк. В данном случае, она возвращает остаток от деления длины строки 
    (len(key)) на размер хеш-таблицы (self.size). Это значение используется как индекс в таблице.'''
    def hash_function(self, key):
        return len(key) % self.size
    '''Функция, которая определяет метод для вставки пары ключ-значение в хеш-таблицу. Он вычисляет хеш для ключа, определяет индекс 
    в таблице и добавляет пару в соответствующую ячейку. Если ячейка уже содержит данные (не None), 
    то добавляется вложенный список для обработки коллизий.'''

    def insert(self, key, value):
        index = self.hash_function(key)

        if self.table[index] is None:
            # Если ячейка пуста, создаем связанный список
            self.table[index] = [(key, value)]
        else:
            # Если ячейка не пуста, добавляем элемент в связанный список
            self.table[index].append((key, value))
    '''Метод для вывода содержимого хеш-таблицы в виде датасета. 
    Проходит по всем элементам таблицы и выводит индекс ячейки и ее содержимое.'''
    def display_dataset(self):
        for i, cell in enumerate(self.table):
            print(f"Cell {i}:", cell)

# создадим хэш-таблицу размером 6 строк
hash_table = HashTable(size=6)

# Коллизия: два различных ключа ("яблоко" и "ананас") приводят к одному и тому же индексу
hash_table.insert("яблоко", 5)
hash_table.insert("апельсин", 8)
hash_table.insert("груша", 12)
hash_table.insert("абрикос", 15)
hash_table.insert("киви", 20)
hash_table.insert("ананас", 7)

# Вывод значений в виде датасета
hash_table.display_dataset()
```
