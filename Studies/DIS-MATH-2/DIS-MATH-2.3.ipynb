### Юнит 3. 

Задача 3.4.  
Какая из следующих формул наиболее близко описывает структуру предложения: «Если первый сотрудник не присутствует на работе, тогда если и второй сотрудник отсутствует, то необходимо позвонить начальнику»? x — значит, что первый работник отсутствует, y — значит, что второй работник отсутствует, z — значит, что необходимо позвонить начальнику.

По условию можно сделать вывод, что в необходимой формуле будет 2 импликации, так как условия 2. Так отсеивается 2 формулы из возможных вариантов ответа. К тому же, формулы с ^ нам не подъодят, так как у нас условие в предложении последовательное, а не конъюнктивное. Формула (x -> y) -> z не подходит, так формально значит "Если верно, что (из отсутствия первого следует отсутствие второго), то нужно позвонить начальнику". Это отличается от заданного предложения. Тогда нужная формула будет иметь вид: x -> (y -> z). По человечески: если первый сотрудник (х) отсутствует, тогда (и только тогда) проверяем наличие второго (y) и если он тоже отсутствует, звоним начальнику (z). 

Задача 3.10.

Обозначим через x ↓ y (стрелка Пирса) функцию такую, что она принимает 1 в случае, когда оба аргумента равны 0 (иначе значение равно 1). Соедините эквивалентные формулы (задающие одну и ту же функцию алгебры логики).

Для решения нужно, чтобы выполнялось условие истинности для условия "Вариант задачи ^ вариант ответа".
Высказывание со стрелкой Пирса истинно только тогда, когда оба условия (операнда) ложны (оба равны 0).
Соединив формулы с вариантами ответа, получим:  
A : ((x↓x)↓(y↓y) ≡ x^y) - слева отрицаются отрицательные значения x и y, тогда появляется их конъюнкция;  
B : (((x↓x)↓y)↓((x↓x)↓y) ≡ x->y) - условие слева ложно только тогда, когда ложно y, что соответсвует импликации, которая у нас справа;  
C : ((x↓x) ≡ !(x)) - это выражение следует из определения стрелки Пирса. Также можно добавить в него (≡!(x∨x));  
D : ((x↓y)↓(x↓y) ≡ x∨y) - чтобы условие слева было истинно, x или y должны быть ложны, что соответствует дизъюнкции x или y.

Задача 4.2.  
Дана булева формула x1->(x2->(...(x10))), то есть 9 вложенных импликаций. Сколько имеется значений, на которых эта формула принимает значение 0?

Данная цепочка состоит из ипликаций, что делает её полностью зависимой. В таком случае (если вспоминать присказку с менеджментом и коллективом из курса) условие выполняется, только если ВСЕ сотрудники отсутствуют. Следовательно, такой исход у нас только 1. Это и будет ответом.

Задача 4.3.  
На вход программа получает 4 булевых значения. Сколько различных входных значений существует для этой программы? Программы будем называть различными, если существует значение входных переменных, на которых эти программы дали разные значения.  

Стандартная задача комбинаторики. Булевых значений может быть всего 2 - 0 и 1 соответственно. Всего значений - 4. Собственно, получаем ответ, вспоминая для бинарных выражений - 2^n. Подставляем n=4 и получаем ответ 16.

Задача 4.4.  
На вход программа получает 3 булевых значений, а на выход выдаёт 0 или 1 в зависимости от данных значений. Сколько различных программ такого вида существует? Программы будем называть различными, если существует значение входных переменных, на которых эти программы дали разные значения.

Усложненная версия предыдущей задачи. У нас имеется 3 булевых значения для входа, и столько же для выхода. По той же формуле посчитаем возможное количество комбинаций для входа (или выхода) по уже знакомой формуле: 2**n, где в нашем случае n=3. Получаем общее число комбинаций для действия - 2^3 = 8. Так как у нас имеется вход и выход программы, можно сказать, что у нас ещё одно бинарное выражение, для которого возможно 8 исходов. То есть, используем всё ту же формулу, но n=8 в данном случае. То есть 2^8 = 256. Что и является ответом.
