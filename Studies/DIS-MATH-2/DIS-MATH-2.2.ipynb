### Юнит 2. Задачи на понятия и правила

Задание 2.8: Пусть А = {1, 2, 3}. Чему равно P(A)?

От нас требуется найти мощность множества. Помним, что мощность состоит из всех возможных сочетаний множества, включая пустое множество. А по правилу мощности она равна $2^n$, где n-количество элементов множества. Посчитаем, получим 8 элементов. Перечислим все возможные сочетания: {$\varnothing$,\{1\},\{2\},\{3\},\{1,2\},\{2,3\},\{1,3\},\{1,2,3\}\}. Итого 8 элементов. Всё сошлось.

Задание 2.10: Перечислите через запятую в порядке возрастания элементы множества {a ∈ R | $a^2$ = 1}.

Элементы множества принадлежит к действительным числам, при этом задаются уравнением $a^2$ = 1. Решая простое уравнение, получаем, что множество состоит из чисел {-1,1}.

Задание 3.2: Пусть A = {3, 6, 7}, перечислите через запятую в порядке возрастания элементы множества {n ∈ Z | n + 2 ∈ A}.

Для выполения задания нужно найти числа, удовлетворяющие условию: целые числа, которые равны (3, 6, 7), которые получаются в результате функции n + 2. Решая легкое уравнение вида n + 2 = 3 (для первого числа), получаем, что n1 = 1. Аналогично получаем остальные числа. Итоговое искомое множество: {1, 4, 5}.

Задание 3.4: Известно, что на собеседование пришли кандидаты со следующими свойствами:  
- ранее не работал в крупной компании
- умеет программировать на Python
- не знает SQL  

Также известно, что на второе собеседование были приглашены следующие кандидаты: обладающие всеми тремя свойствами; не обладающие первым или третьим, но обладающие вторым; обладающие вторым, но не обладающие первым. Кого в итоге пригласили?

Для начала определимся с положительными, отрицательными и нейтральными чертами. Работал или нет человек в крупной компании - это скорее нейтральное свойство, так как в целом мы ищем специалиста, и, если это и важно, то явно это свойство не первостепенное; умеет программировать на Python - однозначно положительная черта. Не сказано, кого конкретно мы ищем, но судя по всем остальным свойствам, умение важное, а возможно, мы вообще ищем программиста или дата-саентиста. В любом случае, это навык, и лишним он не будет; незнание SQL - свойство противоположное предыдущему. Нет навыка - это скорее минус. Мы скорее будем искать человека, не обладающего данным свойством.

Итак, идеальный кандидат - умеет программировать. Работал ли он в крупной компании, нам не особо важно, но допустим, что это приятный бонус, и знающие SQL, то есть не обладающие 3-м свойством. По условию, ***из всех кандидатов приоритетными будут те, кто умеет программировать***, далее идут те, кто имеет все 3 свойства, так как если у приоритетных минуса может и не быть, у этих он точно будет. Оставшиеся люди либо просто работали в крупной компании (что они там делали - не понятно), или знающие SQL, но не умеющие программировать (что ставит их знание SQL под вопрос, скорее они просто знакомы с ним).

Задание 3.8: Пусть отношение R = {(1,2),(2,5),(1,3),(2,1)}. Выпишите все элементы Dom(R).

Dom(R) - Область определения R. То есть, это множество всех первых элементов упорядоченных пар, входящиех в отношение R. Для того, чтобы найти искомое, выпишем первые элементы: 1,2,1,2. Далее упорядочим их и уберем повторения - так как у нас множество, в котором элементы не повторяются. Получим ответ: (1,2).
