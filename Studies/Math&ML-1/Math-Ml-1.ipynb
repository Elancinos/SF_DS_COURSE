#### Практика. Обратная матрица, определитель, ранг.

```python
# Задание 10.1
A = np.array([[8, 6, 11], [7, 5, 9],[6, 10, 6]])
print(np.linalg.inv(A))
print('-'*50)

v1 = np.array([9, 10, 7, 7, 9])
v2 = np.array([2, 0, 5, 1, 4])
v3 = np.array([4, 0, 0, 4, 1])
v4 = np.array([3, -4, 3, -1, -4])

# Задание 10.2
matrix_system = np.array([v1, v2, v3, v4]).T
print('Ранг совокупной матрицы:', np.linalg.matrix_rank(matrix_system))
print('-'*50)

#Задание 10.3
print(matrix_system)
print('-'*50)

#Задание 10.4
G = matrix_system.T@matrix_system
print('Матрица Грамма:', G)
print('-'*50)

#Задание 10.5
print('Определитель матрицы Грамма:', round(np.linalg.det(G)))
print('-'*50)

#Задание 10.7
print('Обратная матрица матрице Грамма:\n',
      np.round(np.linalg.inv(G), 3))
```
