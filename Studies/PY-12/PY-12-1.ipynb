```python
#Импортируем нужные библиотеки
import pandas as pd
```

```python
#Читаем файл (Изменённый, но не собственный)

melb_df = pd.read_csv('data/melb_data_fe.csv')
display(melb_df.head())
```

```python
#Повторяем преобразования данных datetime и category

#копия на всякий случай
melb_df_copy = melb_df.copy()

melb_df_copy['Date'] = pd.to_datetime(melb_df_copy['Date'])
print('Число сделок во втором по популярности квартале -', melb_df_copy['Date'].dt.quarter.value_counts().values[1])
# def category_transformation(table):
#     exclude_list = (['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car'])
#     for column in table.columns:
#         if column not in exclude_list and len(table[column].unique()) < 150:
#             table[column] = table[column].astype('category')

for column in melb_df_copy.columns:
    exclude_list = (['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car'])
    if column not in exclude_list and len(melb_df_copy[column].unique()) < 150:
        melb_df_copy[column] = melb_df_copy[column].astype('category')
        
display(melb_df_copy.info()) #category(10)
```

 2. Сортировка данных в DataFrame

```python
display(melb_df_copy.sort_values(by='Price').head())
display(melb_df_copy.sort_values(by='Date', ascending=False).head())
display(melb_df_copy.sort_values(by=['Distance', 'Price']).loc[::10, ['Distance', 'Price']])
mask1 = melb_df_copy['AreaRatio'] < -0.8
mask2 = melb_df_copy['Type'] == 'townhouse'
mask3 = melb_df_copy['SellerG'] == 'McGrath'
display(melb_df_copy[mask1 & mask2 & mask3].sort_values(
    by=['Date', 'AreaRatio'],
    ascending=[True, False],
    ignore_index=True
).loc[:, ['Date', 'AreaRatio']])
#Задания 2.2-2.3
display(melb_df_copy.sort_values(by='AreaRatio', ascending=False, ignore_index=True).loc[1558, ['BuildingArea']])
mask1 = melb_df_copy['Type'] == 'townhouse'
mask2 = melb_df_copy['Rooms'] > 2
display(melb_df_copy[mask1 & mask2].sort_values(by=['Rooms', 'MeanRoomsSquare'], ascending=[True, False], ignore_index=True).loc[18, ['Price']])
```

3. Группировка данных в DataFrame

```python
display(melb_df_copy.groupby(by='Type').mean(numeric_only=True))
display(melb_df_copy.groupby('Type')['Price'].mean())
display(melb_df_copy.groupby('Regionname')['Distance'].min().sort_values(ascending=False))
display(melb_df_copy.groupby('MonthSale')['Price'].agg(
    ['count', 'mean', 'max']
).sort_values(by='count', ascending=False))
display(melb_df_copy.groupby('MonthSale')['Price'].agg('describe'))
display(melb_df_copy.groupby('Regionname')['SellerG'].agg(
    		['nunique', set]))

#задания 3.1-3.3
#1
display(melb_df_copy.groupby('Rooms')['Price'].mean().sort_values( ascending=False)) #7
#2
display(melb_df_copy.groupby('Regionname')['Lattitude'].std().sort_values()) #Western Victoria
#3
mask1 = melb_df_copy['Date'] >= pd.to_datetime('2017-05-01')
mask2 = melb_df_copy['Date'] <= pd.to_datetime('2017-09-01')
display(melb_df_copy[mask1 & mask2].groupby('SellerG')['Price'].agg('sum').sort_values()) #LITTLE
```
