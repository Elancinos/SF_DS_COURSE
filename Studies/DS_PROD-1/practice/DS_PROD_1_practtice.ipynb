```python
# по совету deepseek для обхода ошибки

import pickle
import warnings

# Подавляем предупреждения
warnings.filterwarnings("ignore")

try:
    with open('model.pkl', 'rb') as f:
        model = pickle.load(f)
except AttributeError as e:
    if "tol" in str(e):
        # Попробуем загрузить модель, игнорируя отсутствие tol
        # Создадим свой Unpickler, который будет пропускать этот атрибут
        import io
        import sys

        class CustomUnpickler(pickle.Unpickler):
            def load(self):
                # Переопределяем загрузку объекта
                # Мы не можем легко пропустить атрибут, поэтому попробуем другой подход
                # Вместо этого, мы загрузим объект, а затем установим tol, если его нет
                obj = super().load()
                if not hasattr(obj, 'tol'):
                    obj.tol = 1e-4
                return obj

        with open('model.pkl', 'rb') as f:
            model = CustomUnpickler(f).load()
    else:
        raise

# Проверяем модель
print("Model loaded")
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```
