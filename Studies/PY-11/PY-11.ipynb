```python
import pandas as pd
```

```python
#Импортируем датасет:

melb_data_ps = pd.read_csv('data/melb_data_ps.csv', sep=',')
display(melb_data_ps)
```

```python
#Создадим копию данных для тестировочных преобразований

melb_data_test = melb_data_ps.copy()
display(melb_data_test.head())
```

```python
#Удаление Index и Coordinates
melb_data_test = melb_data_ps.copy()
melb_data_test.drop(['index', 'Coordinates'], axis=1, inplace=True)
display(melb_data_test.head())

#Общее количество комнат в домах
total_rooms = melb_data_test['Rooms'] + melb_data_test['Bathroom'] + melb_data_test['Bedroom']
display('Общее количество комнат:', total_rooms)
#Средняя площадь комнат в доме
melb_data_test['Mean Rooms Square'] = melb_data_test['BuildingArea'] / total_rooms
display('Средняя площадь комнат в доме:', melb_data_test['Mean Rooms Square'])
#Коэффициент соотношения площадей здания и земли в целом
diff_area = melb_data_test['BuildingArea'] - melb_data_test['Landsize']
sum_area = melb_data_test['BuildingArea'] + melb_data_test['Landsize']
melb_data_test['AreaRatio'] = diff_area/sum_area
display('Коэффициент соотношения площадей здания и земли в целом:', melb_data_test['AreaRatio'])

#Задания 2.1-2.4
countries_df = pd.DataFrame({
    'country': ['Англия', 'Канада', 'США', 'Россия', 'Украина', 'Беларусь', 'Казахстан'],
    'population': [56.29, 38.05, 322.28, 146.24, 45.5, 9.5, 17.04],
    'square': [133396, 9984670, 9826630, 17125191, 603628, 207600, 2724902]
})
countries_df2 = countries_df.copy()
countries_df2['density'] = (countries_df2['population'] * 1000000) / countries_df2['square']
print('Средняя плотность насления данных стран:', round(countries_df2['density'].mean(), 2))

#Разбираемся с датой-временем
melb_data_test['Date'] = pd.to_datetime(melb_data_test['Date'], dayfirst=True)
display(melb_data_test['Date'])
years_sold = melb_data_test['Date'].dt.year
print(years_sold)
print('Min year sold:', years_sold.min())
print('Max year sold:', years_sold.max())
print('Mode year sold:', years_sold.mode()[0])
melb_data_test['MonthSale'] = melb_data_test['Date'].dt.month
melb_data_test['MonthSale'].value_counts(normalize=True)*100
delta_days = melb_data_test['Date'] - pd.to_datetime('2016-01-01') 
display(delta_days.dt.days)
melb_data_test['AgeBuilding'] = melb_data_test['Date'].dt.year - melb_data_test['YearBuilt']
display(melb_data_test['AgeBuilding'])
melb_data_test = melb_data_test.drop('YearBuilt', axis=1)
melb_data_test['WeekdaySale'] = melb_data_test['Date'].dt.day_name()
display(melb_data_test)
weekend_count = len(melb_data_test[melb_data_test['WeekdaySale'] == 'Saturday']) + len(melb_data_test[melb_data_test['WeekdaySale'] == 'Sunday'])
print(weekend_count)

# Задания 3.4-3.5
ufo_data = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv', sep=',')
ufo_data_upd = ufo_data.copy()
ufo_data_upd['Time'] = pd.to_datetime(ufo_data_upd['Time'])
ufo_data_upd['Date'] = ufo_data_upd['Time'].dt.date
display(ufo_data_upd)
display(ufo_data_upd['Time'].dt.year.value_counts()) #Получаем ответ "1999"
print('Средний интервал сообщений в штате NV:', round(ufo_data_upd[(ufo_data_upd['State'] == 'NV')]['Date'].diff().dt.days.mean()), 'дней')
```
